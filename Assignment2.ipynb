1 -Discuss string slicing and provide examples.

Ans - String slicing in Python is a technique to extract a portion (substring) from a string using its indices

Using index we can access any character or piece of string from the string.

Basic Syntax:-
String[start:end:step]

```python
#String Slicing Example:-
name="pankaj Anil Wankhede"
print(name[0:10])
print(name[0:10:2])
print(name[-1:-10:-1])
```

6 - Discuss the use cases of tuples and sets in Python programming.

Ans - Use cases of Tuples

1- Representing fixed data:-uples are ideal for representing fixed collections of data where the order matters and elements should not be modified. Examples include:

-Coordinates: (x, y, z)

-RGB color values: (255, 0, 128)

-Database records: (name, age, city)

2 - Function Return Values: When a function needs to return multiple values, tuples provide a convenient way to package them together.

3 - Dictionary Keys: Tuples can be used as dictionary keys because they are immutable and hashable.

```python
#Tuple use cases
#Function return values
def get_user_info():
  name="pankaj"
  age="22"
  city="wakod"
  return name,age,city
user_info=get_user_info()
print(user_info)

#Dictinary keys
my_dict={
    (1,2):"A",
    (3,4):"B"
}
print(my_dict)
```

Ans - Use cases of sets

1 - Removing Duplicates: Sets automatically eliminate duplicate elements, making them useful for cleaning up data

2 - Membership Testing: Checking if an element is present in a set is very efficient.

3 - Mathematical Set Operations: Sets support operations like union, intersection, and difference, which are useful for tasks like:

-Finding common elements between two datasets.

-Identifying unique elements in each dataset.

 -Combining datasets while removing duplicates.

```python
#Use case of sets
#Removing duplicates
Info={1,"pankaj",2,2,"pankaj",80.5,80.5}
print(Info)

#Membership testing
Sambhajinagar={"Nilesh","Harshal","Nadim","Bushan","Wavre","Saurabh"}
Wakod="pankaj"
if Wakod in Sambhajinagar:
  print("I am present")
else:
  print("I am not present")

#set operations
num1={1,3,6,7,2,8,5}
num2={2,6,5,2,8}
print(num1.union(num2))
print(num1.intersection(num2))
print(num1.difference(num2))
print(num1.symmetric_difference(num2))

```

8- Discuss the importance of dictionary keys being immutable and provide example.

Ans -
Dictionary keys in Python must be immutable. This is because the underlying implementation of dictionaries uses hash tables. When you insert a key, Python calculates its hash value to determine where to store the corresponding value in memory.

If keys were mutable and you changed a key after inserting it, its hash value would also change. This would lead to the dictionary looking for the value in the wrong place, effectively breaking the dictionary's functionality

```python
#Example of dictionary keys that can't be change due to the hash value
my_dict={[1,2]:"Abhay",[2,4]:"Rahul"}
print(my_dict)
#we can use tuple as a key because tuple is immutable
my_dict={(1,2):"Abhay",(2,4):"Rahul"}
print(my_dict)
```
