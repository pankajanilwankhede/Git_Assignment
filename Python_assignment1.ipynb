Que 6-  How do conditional statements work in Python? Illustrate with examples.

Ans : - a - if statement :- The if statement evaluates a condition (an expression that returns True or False). If the condition is True, the block of code indented under the if statement is executed.

b- else statement : - The else statement is used to define a block of code to be executed if the condition in the if statement is False.

c - elif statement :- The elif (short for "else if") statement allows you to check multiple expressions for True and execute a block of code as soon as one of the conditions is True. It is used after an if statement and before an else statement

d - Nesting statement :- The block within the if ,elif or else statement.

e - Ternary Operator :- Python also supports a more concise way of writing conditional statements using a conditional expression .


Let's see some example related to the above concept.......

```python
# 1-if statement
age =18
if age >= 18:
    print("You are eligible to vote.")
# 2 -else statement
age = 17
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
# 3 -elif statement
if age >= 18:
    print("You are eligible to vote.")
elif age >= 13:
    print("You are a teenager.")
else:
    print("You are not eligible to vote.")
# 4 -Nesting statement
if age >= 18:
    if age >= 60:
        print("You are a senior citizen.")
    else:
        print("You are an adult.")
else:
    print("You are not eligible to vote.")
# 5 -Ternary operator
mark =50
status ="pass" if mark>=40 else "fail"
print(status)
```

Que 7- Describe the different types of loops in Python and their use cases with examples.

Their are two types of loops in python ;-

a - for loop :- A for loop is used to iterate over a sequence (such as a list, tuple, dictionary, set, or string) or any other iterable object. The loop iterates over each element in the sequence, executing the block of code for each element

use case ❎
Iterating over items in a list, tuple, dictionary, or set

Iterating over characters in a string

Running a block of code a fixed number of times


b - while loop :-A while loop repeatedly executes a block of code as long as the specified condition is True. The condition is evaluated before the execution of the loop's body, and if it is False, the loop terminates.

use case ❎
Repeating a block of code indefinitely until a condition is met

Implementing loops where the number of iterations is not known beforehand

Waiting for a specific condition to become True

```python
#for loop
name ="pankaj"
for n in name:
  print(n)
names =["Rahul","Tushar","Aaba","Didi","Bondu"]
for name in names:
  print(name)
for i in range(1,9):
  print(i)
for i in range(1,9,3):
  print(i)
for i in range(9,1,-2):
  print(i)
for i in range(10):
  print(i)
for i in range(10):
  print(i)
  if i==5:
    break
for i in range(10):
  if i%2==0:
    continue
  print(i)

```
